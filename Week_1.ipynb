```python
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=30,kde=True)
plt.title("POWER GENERATION")
plt.xlabel("Generated Power in kw")
plt.ylabel("Frequency")
plt.show()
```

```python
# SCATTER PLOT
```

```python
plt.figure(figsize=(15,30))
for i, column in enumerate(df.columns):
    plt.subplot(7,3, i+1)
    plt.scatter(df[column], df['generated_power_kw'])
    plt.title(f'{column} vs Genrated Power (KW)')
    plt.xlabel(column)
    plt.ylabel('generated power kw')
plt.tight_layout()
plt.show()
```

```python
df.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), cmap='coolwarm',annot=True)
plt.show()
```

```python
plt.figure(figsize=(15,30))
for i,columns in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    sns.boxplot(df[columns])
plt.show()
```
